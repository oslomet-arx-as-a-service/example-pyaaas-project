```python
import pandas as pd
from pyaaas import ARXaaS, Dataset, AttributeType
```

```python
df = pd.read_csv("../../data/fifa19.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# drop columns containing NaN
df = df.dropna(axis=1, how='any')
```

```python
df.head()
```

```python
df.info()
```

```python
dataset = Dataset.from_pandas(df)
```

```python
dataset.describe()
```

```python
dataset.set_attribute_type(AttributeType.IDENTIFYING, 'ID', 'Name', 'Age', 'Photo', 'Flag', 'Club Logo')
```

```python
dataset.describe()
```

### Create RiskProfile

```python
arxaas = ARXaaS("http://localhost:8080")
```

```python
risk_profile = arxaas.risk_profile(dataset)
```

```python
risk_profile.re_identification_risk
```

### Estimated prosecutor risk is 1.0. The dataset re-Identification risk is high.

### Anonymize

#### Create hierarchies
Before we can attempt to anonymize the dataset we need to create generalization hierarchies for the quasiidentifying attributes.

```python
from pyaaas.hierarchy import OrderHierarchyBuilder, IntervalHierarchyBuilder
```

### Nationality

```python
nationalities = df["Nationality"].unique().tolist()
```

```python
ob = OrderHierarchyBuilder()
```

```python
nationalities_hierarchy = arxaas.hierarchy(ob, nationalities)
```

```python
nationalities_hierarchy[:10]
```

### Overall

```python
df['Overall'].describe()
```

```python
overalls = df['Overall'].tolist()
```

```python
overall_interval = IntervalHierarchyBuilder()
```

```python
overall_interval.add_interval(0,63, 'low')
overall_interval.add_interval(63,80, 'medium')
overall_interval.add_interval(80,100, 'high')
```

```python
overall_hierarchy = arxaas.hierarchy(overall_interval, overalls)
```

```python
overall_hierarchy[:10]
```

### Potential

```python
df['Potential'].describe()
```

```python
potential = df['Potential'].tolist()
```

```python
potential_interval = IntervalHierarchyBuilder()
```

```python
potential_interval.add_interval(0.0,63.0,)
potential_interval.add_interval(63.0,80.0)
potential_interval.add_interval(80.0,100.0)
```

```python
potential_hierarchy = arxaas.hierarchy(potential_interval, potential )
```

```python
potential_hierarchy[:10]
```

### Value

```python
values = df["Value"].unique().tolist()
```

```python
ob = OrderHierarchyBuilder()
```

```python
values_hierarchy = arxaas.hierarchy(ob, values)
```

```python
values_hierarchy[:10]
```

### Wage

```python
wages = df["Wage"].unique().tolist()
```

### Special

```python
special = df["Special"].unique().tolist()
```

```python
special_hierarchy 
```

### Set hierarchies

```python
dataset.set_hierarchy("Nationality", nationalities_hierarchy)
```

```python
dataset.set_hierarchy("Overall", overall_hierarchy)
```

```python
dataset.set_hierarchy("Potential", potential_hierarchy)
```

```python
dataset.set_hierarchy("Value", values_hierarchy)
```

```python
dataset.set_hierarchy("Wage", wages_hierarchy)
```

```python
dataset.set_hierarchy("Special", special_hierarchy)
```

```python
from pyaaas.privacy_models import KAnonymity
```

```python
kanon = KAnonymity(4)
```

### Attempt anonymization

```python
anon_result = arxaas.anonymize(dataset, [kanon]) 
```

```python
anon_result.dataset.to_dataframe().head()
```

```python
anon_metrics = anon_result.anonymization_metrics
```

```python
anon_metrics.attribute_generalization
```

### New re-identification risk

```python
anon_risk_profile = anon_result.risk_profile
```

```python
anon_risk_profile.re_identification_risk
```

```python
before_risk = risk_profile.re_identification_risk_dataframe()[['estimated_prosecutor_risk', 'highest_prosecutor_risk', 'lowest_risk']].copy()
before_risk["from"] = "before"
```

```python
after_risk = anon_risk_profile.re_identification_risk_dataframe()[['estimated_prosecutor_risk', 'highest_prosecutor_risk', 'lowest_risk']].copy()
after_risk["from"] = "after"
```

```python
risks_df = pd.concat([before_risk, after_risk])
```

```python
risks_df.plot(kind="bar", x="from", title="Re-Identification risks before and after anonymization", figsize=(10, 10));
```
